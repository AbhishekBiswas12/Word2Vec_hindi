#### Removing punctuations and wierd text

```python
import regex as re

# Step 1: Remove ALL Unicode punctuation (\p{P}) and symbols (\p{S})
cleaned_text = re.sub(r'[\p{P}\p{S}]', '', data)

# Step 2: Explicitly remove all Devanagari Digits using their Unicode range
# Devanagari digits range from U+0966 (०) to U+096F (९)
cleaned_text = re.sub(r'[\u0966-\u096F]', '', cleaned_text)

# Step 3: Keep only Devanagari characters (now only letters as digits were removed in Step 2),
# English Digits, and Whitespace.
# Note: \p{Script=Devanagari} here will effectively match Devanagari letters after Step 2.
pattern_final_filter = r'[^\p{Script=Devanagari}0-9\s]'
cleaned_text = re.sub(pattern_final_filter, '', cleaned_text)

# Normalize whitespace (collapse multiple spaces into one) and strip leading/trailing spaces
cleaned_text = re.sub(r'\s+', ' ', cleaned_text).strip()

# Add space between Devanagari character (which should now be only letters) and an English digit, and vice-versa
# We use \p{Script=Devanagari} here, as by this stage it refers only to letters.
cleaned_text = re.sub(r'(\p{Script=Devanagari})(\d)', r'\1 \2', cleaned_text)
cleaned_text = re.sub(r'(\d)(\p{Script=Devanagari})', r'\1 \2', cleaned_text)

# Final whitespace normalization after adding spaces
cleaned_text = re.sub(r'\s+', ' ', cleaned_text).strip()
```

```python
from collections import defaultdict

def pair_freq(vocab):
  pairs = defaultdict(int)
  for word, freq in vocab.items():
    symbols = word.split()
    for i in range(len(symbols) - 1):
      pairs[(symbols[i], symbols[i + 1])] += freq
  return pairs
```

```python
def merge_common(best, vocab):
  bigram = " ".join(best)
  replacement = "".join(best)
  new_vocab = {}
  for word in vocab:
    new_word = word.replace(bigram, replacement)
    new_vocab[new_word] = vocab[word]
  return new_vocab
```

```python
num_merges = 100

for i in range(num_merges):
  pairs = pair_freq(vocab)
  # Find most frequent pair
  best = max(pairs, key=pairs.get)

  print(f"\nMerge {i+1}: {best} -> {''.join(best)}")

  # Merge pair
  vocab = merge_common(best, vocab)

  # Print updated vocab
  print(f"Iteration {i+1} completed!")
```

Finding the current positive predictions for each context word

```python
import numpy as np
import pandas as pd
import time
import os # Import os for file operations

vocab_size = len(unique_words)
oversample_factor = 2

# Define the output file name for negative samples chunks
neg_samples_chunk_file = 'neg_sg_pairs_chunk_temp.csv'

# Remove the file if it already exists to start fresh, and write header
if os.path.exists(neg_samples_chunk_file):
    os.remove(neg_samples_chunk_file)
# Write header initially (this ensures the CSV is properly structured for appending)
pd.DataFrame(columns=['int', 'target', 'pred']).to_csv(neg_samples_chunk_file, mode='w', header=True, index=False)

processed_context_words_count = 0
start_time = time.time()

for context_word_id, (positive_targets, num_positive_samples) in tar_position.items():
    forbidden_words_set = set(positive_targets)
    forbidden_words_set.add(context_word_id)

    collected_negative_samples_for_current_word = []

    while len(collected_negative_samples_for_current_word) < num_positive_samples:
        needed_count = num_positive_samples - len(collected_negative_samples_for_current_word)
        batch_size = needed_count * oversample_factor

        candidates = np.random.randint(0, vocab_size, size=batch_size)
        valid_candidates = candidates[~np.isin(candidates, list(forbidden_words_set))]
        collected_negative_samples_for_current_word.extend(valid_candidates)

    final_negative_samples = collected_negative_samples_for_current_word[:num_positive_samples]

    # Create a small DataFrame for the current context word's negative samples
    temp_df_chunk = pd.DataFrame({
        'int': [context_word_id] * len(final_negative_samples),
        'target': final_negative_samples,
        'pred': 0
    })

    # Append the chunk to the temporary CSV file without writing header again
    temp_df_chunk.to_csv(neg_samples_chunk_file, mode='a', header=False, index=False)

    processed_context_words_count += 1
    if processed_context_words_count % 1000 == 0:
        elapsed_time = time.time() - start_time
        print(f"Processed {processed_context_words_count}/{len(tar_position)} context words. Elapsed: {elapsed_time:.2f}s")

end_time = time.time()
print(f"Total time taken for negative sampling and writing to chunk file: {end_time - start_time:.2f}s")
```
